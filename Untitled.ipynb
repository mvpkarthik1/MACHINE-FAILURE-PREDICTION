```python
import streamlit as st
import pandas as pd
import pickle
#---------------------------------#

pickle_in = open("RF_model.pkl","rb")
RF_model = pickle.load(pickle_in)

#---------------------------------#
# Page layout
## Page expands to full width
st.set_page_config(page_title='MACHINE FAILURE PREDICTION',
    layout='wide')

#---------------------------------#   
st.header("""
MACHINE FAILURE PREDICTION

Enter the values in the input boxes and check the machine failure prediction

""")

#---------------------------------#
st.markdown('**ENTER THE VALUES**')
Type = (st.text_input("Value","enter Type here")
Air_temp = st.text_input("Value","enter Air_temp here")
Process_temp = st.text_input("Value","enter Process_temp here")
Rot_speed = st.text_input("Value","enter Rot_speed here")
Torque = st.text_input("Value","enter Torque here")
Tool_wear = st.text_input("Value","enter Tool_wear here")

Type = int(Type)
Air_temp =  float(Air_temp)
Process_temp = float(Process_temp)
Rot_speed = int(Rot_speed)
Torque = float(Torque)
Tool_wear = int(Tool_wear)
        
X = pd.DataFrame(Type,Air_temp,Process_temp,Rot_speed,Torque,Tool_wear)

if st.button('PREDICT MACHINE FAILURE'):
    predict = RF_model.predict(X)
        
        
    TWF = predict([0])
    HDF = predict([1])
    PWF = predict([2])
    OSF = predict([3])
    RNF = predict([4])
    Mach_failure = predict([5])
            
    if TWF==1 :
            st.write("Tool Wear Failure Occured")
    else TWF==0 :
            st.write("Tool Wear Failure  didnt Occur")

    if HDF==1 :
            st.write("Heat Dissipation Failure Occured")
    else HDF==0 :
            st.write("Heat Dissipation Failure  didnt Occur")
        
        
    if PWF==1 :
            st.write("Power Failure Occured")
    else PWF==0 :
            st.write("Power Failure  didnt Occur")
        
        
    if OSF==1 :
            st.write("Overstrain Failure Occured")
    else OSF==0 :
            st.write("Overstrain Failure  didnt Occur")
        
        
    if RNF==1 :
            st.write("Random Failure Occured")
    else RNF==0 :
            st.write("Random Failure  didnt Occur")
        
        
    if Mach_failure==1 :
            st.write("Machine Failure Occured")
    else Mach_failure==0 :
            st.write("Machine Failure  didnt Occur")
        
```
