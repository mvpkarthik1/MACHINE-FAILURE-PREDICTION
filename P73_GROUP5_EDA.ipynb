# MODEL BUILDING

# LogisticRegression


```python
from sklearn.multioutput import MultiOutputRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score
```

```python
X_train
```

```python
Y_train
```

```python
log_reg = LogisticRegression(max_iter=1000)
model_log_reg = MultiOutputRegressor(log_reg)
model_log_reg.fit(X_train,Y_train)
```

```python
# make predictions for test data
log_reg_y_pred = model_log_reg.predict(X_test)

# evaluate predictions
acc_log_reg = accuracy_score(Y_test, log_reg_y_pred)
print("Accuracy: %.2f%%" % (acc_log_reg * 100.0))
```

```python
print(classification_report(Y_test,log_reg_y_pred))
classification_report(Y_test,log_reg_y_pred)
```

```python
R2_log_reg = r2_score(Y_test, log_reg_y_pred)
R2_adj_log_reg = 1 - (1-model_log_reg.score(X_train, Y_train))*(len(Y_train)-1)/(len(Y_train)-X_train.shape[1]-1)
print(R2_log_reg,R2_adj_log_reg)
```

# KNN

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_KNN = KNeighborsClassifier(n_neighbors=3)
KNN_model = MultiOutputRegressor(model_KNN)
KNN_model.fit(X_train,Y_train)
```

```python
# make predictions for test data
KNN_y_pred = KNN_model.predict(X_test)

# evaluate predictions
acc_KNN = accuracy_score(Y_test, KNN_y_pred)
print("Accuracy: %.2f%%" % (acc_KNN * 100.0))
```

```python
print(classification_report(Y_test,KNN_y_pred))
```

```python
R2_KNN = r2_score(Y_test, KNN_y_pred)
R2_adj_KNN = 1 - (1-KNN_model.score(X_train, Y_train))*(len(Y_train)-1)/(len(Y_train)-X_train.shape[1]-1)
print(R2_KNN,R2_adj_KNN)
```
